```python
import os
import pickle
import pandas as pd

# Assuming all your pickle files are in the './datasets/' directory
input_folder = './datasets_pkl'
output_folder = './datasets_csv'  # You can change this to your desired output folder

# Create the output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Create empty lists to accumulate data
rows = []

# Iterate over each pickle file in the input folder
for file_name in os.listdir(input_folder):
    if file_name.endswith('.pkl'):
        # Load the pickle file
        file_path = os.path.join(input_folder, file_name)
        if file_path == "./sents_borges.pkl":
            continue
        with open(file_path, 'rb') as f:
            data = pickle.load(f)
        print(file_name)
        print(data)
        print(type(data))
        output_file_name = file_name.split(".")[0]+".csv"
        output_file_path = output_folder+"/"+output_file_name
        data.to_csv(output_file_path, index=False)
        print(f"Converted and saved to {output_file_path}")
print("Finished")
```
